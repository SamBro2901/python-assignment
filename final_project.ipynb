```python
# Step 3: Extract the faces from the images and paste into contact sheet

from PIL import Image, ImageDraw

import cv2 as cv
import numpy as np
import re

nfiles = 14

# Task 3: Identify the face pics in the photos

face_cascade = cv.CascadeClassifier('readonly/haarcascade_frontalface_default.xml')
eye_cascade = cv.CascadeClassifier('readonly/haarcascade_eye.xml')

for i in range(nfiles):
    
    # fetch the png file from directory
    path = "a-"+str(i)+".png"
    print('opening file: ', path)
    
    # convert file to greyscale
    img = cv.imread(path)
    # And we'll convert it to grayscale using the cvtColor image
    gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
    pil_img = Image.fromarray(gray,mode="L")
    #display(pil_img)
    
    # use cs to recognize faces
    faces = face_cascade.detectMultiScale(np.array(pil_img), 1.35, minNeighbors=6)
    print("Finished analyzing the photo")
    
    # open image in color and draw boxes 
    if len(faces) != 0:
    
        rec=faces.tolist()
        color_image=Image.open(path).convert('RGB')
        #drawing=ImageDraw.Draw(color_image)
        #for box in rec:
        #    drawing.rectangle((box[0],box[1],box[0]+box[2],box[1]+box[3]), outline="white")
        #display(color_image)
    
        # copy the face to a separate file 
        faces = []
        for box in rec:
            print('Copying face of box ', box)
            box_coord = (box[0],box[1],box[0]+box[2],box[1]+box[3])
            facepics = color_image.crop(box_coord)
            faces.append(facepics.resize((200,200)))
    
        first_face = faces[0]
        contact_sheet=Image.new("RGB", (first_face.width*5,first_face.height*2))
        x=0
        y=0
    
        for face in faces:
            # Lets paste the current image into the contact sheet
            contact_sheet.paste(face, (x, y) )
            # Now we update our X position. If it is going to be the width of the image, then we set it to 0
            # and update Y as well to point to the next "line" of the contact sheet.
            if x+first_face.width == contact_sheet.width:
                x=0
                y=y+first_face.height 
            else:
                x=x+first_face.width
        
        contact_sheet = contact_sheet.resize((int(contact_sheet.width/2),int(contact_sheet.height/2) ))
        display(contact_sheet)
    
        # output boxes around the recognized faces for specific files
        #box_filename = "boxes-"+str(i)+".dat"
        #with open(box_filename,"w") as f:
        #    for box in rec:
        #        f.write(box)
        #        f.close()
            
    
        filename = "facepics-"+str(i)+".png"
  
        # Using cv2.imwrite() method 
        # Saving the image 
        #cv.imwrite(filename, np.array(contact_sheet))
        contact_sheet.save(filename)
    
    else:
        
        print("No face found in file ", i)
    
    print("Finish File ", i, "!")
```

```python
# Step 4: Inquite the name and return the faces

from PIL import Image, ImageDraw

nfiles = 14

# read all preprocessed data
images = {}
wordlists = []

for i in range(nfiles):
    
    # fetch the png file from directory
    try:
        path = "facepics-"+str(i)+".png"
        img = Image.open(path).convert("RGB")
        images[i] = img
    except:
        print("No " + path + "is found!")
    
    # then read the face files
    path = "a-"+str(i)+"-str.dat"
    with open(path,"r") as f:
        wordlists.append(f.readlines())
        f.close()
        
#print(wordlists[0])
        
# now start the query
name = "Christopher"

print("Now find Christopher")
found_flag = [False for i in range(nfiles)]
for i in range(nfiles):
    for strings in wordlists[i]:
        if name in strings:
            found_flag[i] = True
            
for i in range(nfiles):
    if found_flag[i] == True:
        print("Results found in file a-"+str(i)+".png")
        if i in images:
            display(images[i])
        else:
            print("But there were no faces in that file.")
        
# second query
name = "Mark"

print("Now find Mark")
found_flag = [False for i in range(nfiles)]
for i in range(nfiles):
    for strings in wordlists[i]:
        if name in strings:
            found_flag[i] = True
            
for i in range(nfiles):
    if found_flag[i] == True:
        print("Results found in file a-"+str(i)+".png")
        if i in images:
            display(images[i])
        else:
            print("But there were no faces in that file.")
```
